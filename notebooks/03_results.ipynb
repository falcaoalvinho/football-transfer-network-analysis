# Resultados

O mercado de transferências, quando modelado como grafo direcionado, revela uma estrutura marcadamente 
assimétrica: poucos clubes concentram a maior parte das conexões, e essa concentração reflete com 
precisão as diferenças socioeconômicas entre as regiões analisadas. Ligas inseridas em contextos 
economicamente mais robustos acumulam os maiores volumes de transferências, os vértices de maior 
centralidade e as métricas mais expressivas — padrão compatível com o funcionamento real do mercado.

Esse contraste também validou a classificação proposta pelo projeto. Como esperado, os clubes 
apresentam papéis funcionais bem definidos dentro da rede:

- **Exportadores (~5,43% dos clubes):** alta taxa de saída de jogadores, frequentemente localizados 
  em mercados periféricos ou com forte tradição formadora
- **Importadores (~6,07% dos clubes):** alta taxa de contratações, concentrados em mercados com 
  maior capacidade de investimento
- **Intermediários:** atuam como pontes entre exportadores e importadores, com graus de entrada e 
  saída elevados e centralidade de intermediação acima da média — o grupo funcionalmente mais 
  complexo dos três

**Importadores:** San Diego FC, LAFC, Houston Dynamo FC, CF Montreal e Nashville SC

**Intermediários:** Chelsea, Aston Villa, Brighton & Hove Albion, AC Milan e Manchester City

**Exportadores:** Nottingham Forest, Liverpool, Ajax, Arsenal e Lyon

A distribuição geográfica desses perfis revela fluxos claros: jogadores originados em regiões 
periféricas — frequentemente países do sul global — migram predominantemente em direção à Europa. 
Há, porém, uma tendência de reconfiguração desse fluxo: a América do Norte emerge como destino 
competitivo, impulsionada por investimentos crescentes e pela chegada recente de atletas de alto 
prestígio à região.

A detecção de comunidades reforça essa leitura estrutural. Com Q = 0,597 e 15 comunidades 
identificadas pelo método de Louvain, o grafo exibe estrutura comunitária forte — valores acima de 
0,3 já são considerados significativos na literatura de redes complexas. A assortatividade negativa 
(-0,277) indica que os clubes tendem a se conectar com parceiros de perfis distintos, o que é 
coerente com a lógica do mercado: exportadores se relacionam com importadores, não entre si. As 
comunidades identificadas são coerentes com divisões geográficas e competitivas, sugerindo que os 
fluxos de jogadores ainda obedecem a lógicas regionais mesmo num mercado globalizado.

## Conclusão

A modelagem das transferências como grafo direcionado mostrou-se eficaz para representar a 
complexidade estrutural do mercado futebolístico. A partir de dados reais e métricas clássicas de 
teoria de redes, foi possível identificar perfis funcionais distintos entre os clubes, detectar 
comunidades com coerência geográfica e quantificar as assimetrias que estruturam o sistema global 
de transferências.

Os resultados confirmam que poucos clubes concentram os principais fluxos — localizados 
predominantemente na Europa Ocidental e, crescentemente, na América do Norte — enquanto a maioria 
ocupa posições periféricas, atuando como origem ou destino pontual de jogadores. Essa hierarquia 
reflete desigualdades econômicas reais e sugere que a análise de redes pode ser uma ferramenta 
relevante para estudos mais amplos sobre mobilidade de atletas, estratégias de formação e dinâmicas 
de investimento no futebol profissional.

A abordagem é facilmente extensível: incorporar séries temporais, dados financeiros granulares ou 
atributos contextuais dos clubes abriria espaço para análises com maior poder explicativo — um 
caminho natural para trabalhos futuros a partir desta base.
